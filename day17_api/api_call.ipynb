```python
df
format_url = "https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&language=en-US&page={}&sort_by=popularity.desc"
```

```python
import pandas as pd
import requests
import time
dfs = []

for i in range(1, 501):
    print(format_url.format(i))
    # continue
    response = requests.get(format_url.format(i), headers=headers)
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    dfs.append(temp_df)
    time.sleep(5)

# Concatenate all DataFrames at once
df = pd.concat(dfs, ignore_index=True)
```
